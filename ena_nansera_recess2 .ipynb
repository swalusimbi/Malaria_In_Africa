## Handling Null values

```python
# filling Null values with zero
columns_to_fill_with_zeros = ['Use_of_insecticide-treated_bed_nets_(%_of_under-5_population)','People_using_safely_managed_sanitation_services_(%_of_population)']
df1[columns_to_fill_with_zeros] = df1[columns_to_fill_with_zeros].fillna( value = 0)
print()
```

```python
df1.isnull().sum()
```

```python

# Choosing the column names with missing values to fill with the mean
columns_to_fill_mean = ['Rural population (% of total population)','Urban population (% of total population)','People using at least basic sanitation services (% of population)','Incidence of malaria (per 1,000 population at risk)','Malaria cases reported']

mean_values = df1[columns_to_fill_mean].mean()

# Fill missing values in each column with its respective mean
for column in columns_to_fill_mean:
    df1[column].fillna(mean_values[column], inplace=True)

```

```python
#  Group by year and count the number of Malaria Incidences for each year
yearly_incidence_count = df.groupby('Year')['Incidence_of_malaria_(per_1,000_population_at_risk)'].count().reset_index()

# Print the yearly Incidence count
print(yearly_incidence_count)
```
